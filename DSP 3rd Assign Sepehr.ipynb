```python
import urllib.request
import librosa
from IPython import display
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Download an example audio file
file_name = "2776.mp3"  # Source: https://bigsoundbank.com/detail-2776-cockatiel-parakeet-8.html
url = f"https://bigsoundbank.com/UPLOAD/mp3/{file_name}"

urllib.request.urlretrieve(url, file_name)
sound_data, samplerate = librosa.load(file_name)  # sr=None keeps the original sample rate
print(samplerate)

display.display(display.Audio(data=sound_data, rate=samplerate))
```

```python
# Plot the time domain signal
import matplotlib.pyplot as plt

times = np.arange(len(sound_data)) / float(samplerate)
plt.figure(figsize=(10, 3))
plt.plot(times, sound_data)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform (time domain)")
plt.tight_layout()
plt.show()
```

```python
# Transform to frequency domain using FFT
from scipy.fft import fft, fftfreq

x = sound_data
fs = samplerate
N = len(x)


X = fft(x)
freqs = fftfreq(N, d=1/fs) 
magnitude = np.abs(X)

plt.figure(figsize=(10, 4))

plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.plot(freqs, magnitude)
plt.title("Magnitude Spectrum")
plt.xlim(0, fs/2)  
plt.tight_layout()
plt.show()
```

```python
# Download second audio file
file_name2 = "french-horn-signal-3-97844.mp3"  # Source: https://cdn.pixabay.com/download/audio/2022/03/20/audio_d3e142b83c.mp3?filename=french-horn-26.mp3

sound_data2, samplerate2 = librosa.load(file_name2)  
print(samplerate2)
display.display(display.Audio(data=sound_data2, rate=samplerate2))
```

```python
# Plot the time domain of the second audio file
times2 = np.arange(len(sound_data2)) / float(samplerate2)
plt.figure(figsize=(10, 3))
plt.plot(times2, sound_data2)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform (time domain) - French Horn")
plt.tight_layout()
plt.show()
```

```python
#transform second audio to frequency domain
#transform
x2 = sound_data2
fs2 = samplerate2
N2 = len(x2)
X2 = fft(x2)
freqs2 = fftfreq(N2, d=1/fs2)[:N2//2]  # only positive frequencies
magnitude2 = np.abs(X2)
plt.figure(figsize=(10, 4))
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.plot(freqs2, magnitude2[:N2//2])  # plot only positive frequencies
plt.title("Magnitude Spectrum - French Horn")
plt.xlim(0, fs2/2) 
plt.tight_layout()
plt.show()

```
